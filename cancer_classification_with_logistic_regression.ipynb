## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load the dataset

```python
df=pd.read_csv('cancer_data.csv')
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Exploratory Data Analysis

```python
# Count plot of the target variable
sns.set_style('whitegrid')
sns.countplot(x='diagnosis', data=df)
```

```python
# Convert categorical variable into dummy/indicator variables
# 1 represents malignant(M) and 0 represents benign(B)
df=pd.get_dummies(data=df, columns=['diagnosis'], drop_first=True)
```

```python
# drop irrelevant features
df=df.drop(['id','Unnamed: 32'], axis=1)
```

```python
# HeatMap
plt.subplots(figsize=(20,18))
sns.heatmap(df.corr(), annot=True, linewidths=0.5)
```

```python
X=df.drop('diagnosis_M', axis=1)
y=df['diagnosis_M']
```

## Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Training the Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_r=LogisticRegression()
```

```python
log_r.fit(X_train, y_train)
```

## Predicting Test Data

```python
predictions=log_r.predict(X_test)
```

```python
predictions
```

## Evaluating the model

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')

labels=[0,1]
cmx=confusion_matrix(y_test, predictions, labels)
fig = plt.figure()
ax = fig.add_subplot()
cax = ax.matshow(cmx)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print('\n')
print(classification_report(y_test, predictions))
```
